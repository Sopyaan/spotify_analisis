## Visualization & Explanatory Analysis

### Bagaimana hubungan antara music features dan jumlah streams?


```python
# Select relevant columns for analysis
music_features = spotify[['streams','valence_%', 'danceability_%']]

# Pairplot for scatter plots
sns.pairplot(music_features)
plt.suptitle('Scatter Plot Matrix of Music Features and Streams', y=1.02)
plt.show()
```

```python
# Heatmap for correlation matrix
correlation_matrix = music_features.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Music Features and Streams')
plt.show()
```

### Siapa saja artis top 10 dengan lagu terbanyak?


```python
spotify.rename(columns={'artist(s)_name':"artists"}, inplace=True)
```

```python
# Turning the artists column into list instead of a single string
artists = []

for row in spotify['artists']:
    item = row.split(',')
    artists.append(item)

spotify['artists'] = artists
spotify.head()
```

Ratakan kolom artis sekarang untuk mendapatkan satu daftar yang berisi semua nama di kolom artis, meskipun berisi beberapa artis untuk setiap lagu

```python
artists_flattened = []
for item in artists:
    artists_flattened.extend(item)
```

```python
artists_true_counts = pd.Series(artists_flattened).value_counts()
```

```python
#Top 10 artis lagu terbanyak
top_artists = artists_true_counts.head(10)

sns.barplot(x=top_artists.values, y=top_artists.index, palette='crest')
plt.title("Top 10 artists with most songs 2023")
plt.xlabel("Number of songs")
plt.ylabel("Artist Name")

top_artists

```

### Apa lagu paling populer yang dirilis pada tahun 2023?

### Apakah terdapat pola tertentu dalam karakteristik musik yang populer di kalangan pendengar?

```python
# Convert 'streams' column to numeric
spotify['streams'] = pd.to_numeric(spotify['streams'], errors='coerce')

# Specify the popularity threshold
popularity_threshold = 50000000

# Create a new column 'popularity' based on the threshold
spotify['popularity'] = spotify['streams'] >= popularity_threshold
```

```python
# Verify the DataFrame after creating the 'popularity' column
print(spotify.head())  # Check the first few rows to confirm the 'popularity' column is present
```

```python
# Select relevant columns for analysis
music_features = spotify[['streams','valence_%', 'danceability_%']]

# Select relevant columns for analysis
music_features = spotify[['popularity', 'valence_%', 'danceability_%']]
```

```python
# Pairplot for scatter plots
sns.pairplot(music_features, hue='popularity', markers=['o', 's'], palette={True: 'b', False: 'r'})
plt.suptitle('Scatter Plot Matrix of Music Features by Popularity', y=1.02)
plt.show()
```
